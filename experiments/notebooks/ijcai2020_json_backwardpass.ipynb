```python
import json
import pprint

pp = pprint.PrettyPrinter(indent=1)
```

```python
filename = 'nielspasta-split-nested.output.json'
newFilename = 'nielspasta-split-nested-filtered.output.json'
```

```python
def removeAssumptionsFromDerivations(derivations, assumptions):
    newDerivations = [elem for elem in derivations if elem not in assumptions]
    return newDerivations 

assumptions=[{'object': '12',
                'pred': 'paid',
                'subject': 'angie',
                'value': True}]
derivations = [{'object': '12',
                'pred': 'paid',
                'subject': 'angie',
                'value': True},
               {'object': '4',
                'pred': 'paid',
                'subject': 'angie',
                'value': False},
               {'object': '8',
                'pred': 'paid',
                'subject': 'angie',
                'value': False},
               {'object': '16',
                'pred': 'paid',
                'subject': 'angie',
                'value': False}]
pp.pprint(removeAssumptionsFromDerivations(derivations=derivations, assumptions=assumptions))
pp.pprint(assumptions)
```

```python
with open(filename) as  jsonFile:
    data = json.load(jsonFile)['output']
    newData = []
    for explanation in data:
        if('nested_explanations' in explanation):
            newExplanation = explanation.copy()
            
            nestedExplanations = newExplanation['nested_explanations']
            nestedExplanations = [nestedExplanation for nestedExplanation in nestedExplanations if 'reason_sequence' in nestedExplanation]
            
            for nestedExplanation in nestedExplanations:
                reasonSequence = nestedExplanation['reason_sequence']
                
                for reason in reasonSequence:
                    reason['derivations'] = removeAssumptionsFromDerivations(derivations=reason['derivations'], assumptions=reason['assumptions'])
                
            newExplanation['nested_explanations'] = nestedExplanations    
          
            newData.append(newExplanation)
        else:
            newData.append(explanation)
    
    pp.pprint(newData)
    with open(newFilename,'w') as newJsonFile:
        json.dump(newData,newJsonFile)
        
```
